### Representing Text - Bag of Words and TF-IDF
#### Bag of Words (BoW)

A simple way to represent text as numerical vectors by counting word occurrences and representing each sentence as a vector or word counts

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    "Large language models revolutionize business.",
    "Business applications benefit from AI.",
    "Language models learn from text data."
]

#  CountVectorizer is used to convert a collection of text documents to a matrix of token counts.
#  The output is a sparse matrix where each row represents a document and each column represents a word in the corpus.
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

# Get feature names and the array representation of the matrix
feature_names = vectorizer.get_feature_names_out()

df = pd.DataFrame(X.toarray(), columns=feature_names)
df
```

### Term Frequency-Inverse Document Frequency (TF-IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    "Large language models revolutionize business.",
    "Business applications benefit from AI.",
    "Language models learn from text data."
]

tfidf_vectorizer = TfidfVectorizer()

# Transform the corpus into a document-term matrix
# Each row represents a document in the corpus
# Each column represents a unique word in the corpus
X_tfidf = tfidf_vectorizer.fit_transform(corpus)


df = pd.DataFrame(X_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
df
```

## **Practical Example**
### **Sentiment Analysis of Movie Reviews**


This application will go through the entire text preprocessing pipeline and show how it contributes to a real-world NLP task.
Setup

Dataset: Use a small dataset of movie reviews (positive and negative) - you could use a subset of IMDB reviews or create 10-15 simple examples.

#### Text Processing Pipeline
**Raw Text → Tokenization → Normalization → Stop Words Removal → Stemming/Lemmatization → BoW → Classification**

Next, we will do the following:
- Import the needed libraries
- Download any additional modules that are arequired
- Prepare the dataset i.e Text Corpus

```python
import pandas as pd
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Sample movie reviews dataset
reviews = [
    {"text": "This movie was absolutely fantastic! Great acting and storyline.", "sentiment": 1},
    {"text": "I loved this film. The characters were so well developed.", "sentiment": 1},
    {"text": "Amazing cinematography and directing. One of the best films I've seen.", "sentiment": 1},
    {"text": "The acting was good but the story was too predictable.", "sentiment": 0},
    {"text": "Terrible movie. I wasted two hours of my life.", "sentiment": 0},
    {"text": "The special effects were amazing but everything else was boring.", "sentiment": 0},
    {"text": "I enjoyed the action sequences but the dialogue was poorly written.", "sentiment": 0},
    {"text": "Brilliant performance by the lead actor! Highly recommended.", "sentiment": 1},
    {"text": "So disappointing. The trailer was better than the actual movie.", "sentiment": 0},
    {"text": "A masterpiece of modern cinema. I was captivated throughout.", "sentiment": 1}
]

# Create DataFrame
df = pd.DataFrame(reviews)
print("Original Data:")
print(df.head())
print("\n")
```

In the text segment below, we will define the required text processing functions:
1. tokenize text function
2. normalize_tokens function
3. remove_stopwords function
4. stem_tokens function
5. lemmatize_tokens function
6. preprocess_text function, the pipeline that calls all the previous functions

```python
# Step 1: Tokenization
def tokenize_text(text):
    tokens = word_tokenize(text)
    # print(f"Sentence after tokenization: {tokens}") # uncomment if need to see results
    return tokens

# Step 2: Normalization
def normalize_tokens(tokens):
    # Convert to lowercase and remove punctuation
    normalized = [re.sub(r'[^\w\s]', '', token.lower()) for token in tokens]
    normalized = [token for token in normalized if token]  # Remove empty strings
    # print(f"Tokens after normalization: {normalized}") # uncomment if need to see results
    return normalized

# Step 3: Remove stop words
def remove_stopwords(tokens):
    stop_words = set(stopwords.words('english'))
    filtered = [token for token in tokens if token not in stop_words]
    # print(f"Tokens after stopword removal: {filtered}") # uncomment if need to see results
    return filtered

# Step 4a: Stemming
def stem_tokens(tokens):
    stemmer = PorterStemmer()
    stemmed = [stemmer.stem(token) for token in tokens]
    # print(f"Tokesn after stemming: {stemmed}") # uncomment if need to see results
    return stemmed

# Step 4b: Lemmatization
def lemmatize_tokens(tokens):
    lemmatizer = WordNetLemmatizer()
    lemmatized = [lemmatizer.lemmatize(token) for token in tokens]
    # print(f"Tokens after lemmatization: {lemmatized}") # uncomment if need to see results
    return lemmatized

# Complete preprocessing pipeline
def preprocess_text(text, use_stemming=True):
    tokens = tokenize_text(text)
    normalized_tokens = normalize_tokens(tokens)
    no_stopwords = remove_stopwords(normalized_tokens)
    
    if use_stemming:
        return stem_tokens(no_stopwords)
    else:
        return lemmatize_tokens(no_stopwords)

# Demonstrate the preprocessing pipeline on one example
print("PREPROCESSING PIPELINE DEMONSTRATION:")
sample_text = df['text'][0]
print(f"Original text: '{sample_text}'")
processed_tokens = preprocess_text(sample_text, use_stemming=True)
print(f"Processed tokens: {processed_tokens}")
print("\n")
```

Apply the text processing pipeline on all our sentences (entire text corpus)

```python
df['processed_text'] = df['text'].apply(lambda x: ' '.join(preprocess_text(x, use_stemming=True)))
df
```

Convert text into numerical representation using the Bag of Words Method, we will the use the simple count approach

```python
# Bag of Words representation
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['processed_text'])

print("BOW Matrix (First 3 Documents):")
bow_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
bow_df
```

Create the classifier and train it on the generated BoW representations

```python
X_train, X_test, y_train, y_test = train_test_split(X, df['sentiment'], test_size=0.3, random_state=42)

# Train a Naive Bayes classifier
clf = MultinomialNB()
clf.fit(X_train, y_train)
```

Use the trained classifier to make predictions using the test data (sentences that was not seen by the classifier during training)

```python
y_pred = clf.predict(X_test)
```

Evaluate the classification results (Model Performance)

```python
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

## **Part 2:- Language Models**

### Building a Simple N-gram Language Model

```python
import random
from collections import defaultdict, Counter
import re
import math

class NGramModel:
    def __init__(self, n):
        """
        Initializes the N-gram language model.
        :param n: The 'n' in n-gram. (e.g., 2 for bigram, 3 for trigram)
        """
        self.n = n
        # self.ngrams will map a context (of length n-1) to a Counter of next words
        self.ngrams = defaultdict(Counter)
    
    def tokenize(self, text):
        """
        A simple tokenizer that lowercases the text and separates words and basic punctuation.
        """
        # This simple regex extracts words (including words with apostrophes) and basic punctuation
        tokens = re.findall(r"[\w']+|[.,!?;]", text.lower())
        return tokens

    def train(self, text):
        """
        Trains the model by calculating the frequencies of next words given their preceding context.
        """
        tokens = self.tokenize(text)
        
        # We need at least n tokens to form one valid n-gram
        if len(tokens) < self.n:
            print("Warning: Corpus is too short to train an n-gram model of this size.")
            return
        
        # Slide a window of size 'n' across our tokens
        for i in range(len(tokens) - self.n + 1):
            # The context is the first n-1 words in our window
            context = tuple(tokens[i : i + self.n - 1])
            # The target is the nth word, which follows the context
            next_word = tokens[i + self.n - 1]
            
            # Count the occurrence of next_word given the specific context
            self.ngrams[context][next_word] += 1
            
    def print_dictionary(self):
        """
        Prints the dictionary of contexts and counts clearly, one line per key-value pair.
        """
        print("Dictionary Elements:")
        for context, next_words in self.ngrams.items():
            context_str = " ".join(context)
            if not context_str:
                context_str = "<empty>"
            
            counts_dict_str = "{" + ", ".join([f"'{w}': {c}" for w, c in next_words.items()]) + "}"
            print(f"    '{context_str}' -> {counts_dict_str}")
        print()

    def generate(self, num_words, seed_text=None):
        """
        Generates text of length `num_words`, optionally starting with `seed_text`.
        """
        generated_tokens = []
        context = ()
        
        # Determine the initial context from the seed_text
        if seed_text:
            seed_tokens = self.tokenize(seed_text)
            # We add all the seed tokens to our output
            generated_tokens.extend(seed_tokens)
            
            # The context needs to be exactly of length (n-1) to predict the next word
            if len(seed_tokens) >= self.n - 1:
                # Take the last n-1 tokens from the seed to form the context
                context = tuple(seed_tokens[-(self.n - 1):])
            else:
                print(f"Warning: Seed text has fewer than {self.n - 1} words. Falling back to a random start.")
                context = random.choice(list(self.ngrams.keys()))
                generated_tokens = list(context)
        else:
            # If no seed is provided, pick a random starting context that exists in the text
            context = random.choice(list(self.ngrams.keys()))
            generated_tokens = list(context)
            
        # Start generating new words
        for _ in range(num_words):
            # Get the possible next words and their frequencies for the current context
            next_word_counts = self.ngrams.get(context, None)
            
            # If we've reached a dead end (a context we've never seen before), stop early!
            if not next_word_counts:
                break
            
            # Gather words and their weights (counts)
            possible_words = list(next_word_counts.keys())
            weights = list(next_word_counts.values())
            
            # Pick a word randomly, but weighted by its frequency! 
            # (Words that appeared more often following this context have a higher probability)
            next_word = random.choices(possible_words, weights=weights, k=1)[0]
            
            generated_tokens.append(next_word)
            
            # Shift the context window: drop the oldest word, append the newly generated word
            if self.n > 1:
                context = tuple(list(context[1:]) + [next_word])
                
        # Detokenize slightly to make the text readable (attach punctuation to the previous word)
        output_text = " ".join(generated_tokens)
        output_text = output_text.replace(" .", ".").replace(" ,", ",").replace(" !", "!").replace(" ?", "?")
        return output_text.capitalize()

    def get_vocab(self):
        """Helper to get all unique words seen during training."""
        vocab = set()
        for context, next_words in self.ngrams.items():
            for word in next_words:
                vocab.add(word)
            for word in context:
                vocab.add(word)
        return vocab

    def calculate_perplexity(self, text):
        """
        Calculates the perplexity of the given text based on the trained model.
        A lower perplexity means the model predicts the text better!
        """
        tokens = self.tokenize(text)
        if len(tokens) < self.n:
            print("Text too short to evaluate.")
            return float('inf')
            
        vocab_size = len(self.get_vocab())
        # If vocab is 0, the model hasn't been trained
        if vocab_size == 0:
            return float('inf')
            
        log_prob_sum = 0
        N = len(tokens) - self.n + 1
        
        for i in range(N):
            context = tuple(tokens[i : i + self.n - 1])
            target_word = tokens[i + self.n - 1]
            
            context_counts = self.ngrams.get(context, {})
            total_context_occurrences = sum(context_counts.values())
            target_occurrences = context_counts.get(target_word, 0)
            
            # Laplace Add-1 Smoothing
            # P = (Count(Next Word) + 1) / (Total Context Occurrences + Vocab Size)
            prob = (target_occurrences + 1) / (total_context_occurrences + vocab_size)
            
            log_prob_sum += math.log2(prob)
            
        # Perplexity = 2^(- Average Log Probability )
        avg_log_prob = log_prob_sum / N
        return 2 ** (-avg_log_prob)


# --- Example Usage Sandbox for Students ---
if __name__ == "__main__":
    # Sample 20-line corpus: A short snippet from Alice in Wonderland
    sample_corpus = """
Alice was beginning to get very tired of sitting by her sister on the bank, and of having nothing to do: once or twice she had peeped into the book her sister was reading, but it had no pictures or conversations in it, "and what is the use of a book," thought Alice "without pictures or conversations?"
So she was considering in her own mind (as well as she could, for the hot day made her feel very sleepy and stupid), whether the pleasure of making a daisy-chain would be worth the trouble of getting up and picking the daisies, when suddenly a White Rabbit with pink eyes ran close by her.
There was nothing so very remarkable in that; nor did Alice think it so very much out of the way to hear the Rabbit say to itself, "Oh dear! Oh dear! I shall be late!" (when she thought it over afterwards, it occurred to her that she ought to have wondered at this, but at the time it all seemed quite natural); but when the Rabbit actually took a watch out of its waistcoat-pocket, and looked at it, and then hurried on, Alice started to her feet, for it flashed across her mind that she had never before seen a rabbit with either a waistcoat-pocket, or a watch to take out of it, and burning with curiosity, she ran across the field after it, and fortunately was just in time to see it pop down a large rabbit-hole under the hedge.
In another moment down went Alice after it, never once considering how in the world she was to get out again.
The rabbit-hole went straight on like a tunnel for some way, and then dipped suddenly down, so suddenly that Alice had not a moment to think about stopping herself before she found herself falling down a very deep well.
Either the well was very deep, or she fell very slowly, for she had plenty of time as she went down to look about her and to wonder what was going to happen next.
    """

    print("=== Training and Generating with N-Gram Language Models ===\n")
    print(f"Corpus size: {len(sample_corpus.split())} words.\n")
    
    # 1. Bigram Model (n=2)
    # The context length is n-1 = 1 word
    print("--- 2-Gram (Bigram) Model ---")
    bigram_model = NGramModel(n=2)
    bigram_model.train(sample_corpus)
    bigram_model.print_dictionary()
    generated_bigram = bigram_model.generate(num_words=20, seed_text="Alice")
    print(f"Seed: 'Alice'")
    print(f"Output: {generated_bigram}\n")
    
    # 2. Trigram Model (n=3)
    # The context length is n-1 = 2 words
    print("--- 3-Gram (Trigram) Model ---")
    trigram_model = NGramModel(n=3)
    trigram_model.train(sample_corpus)
    trigram_model.print_dictionary()
    generated_trigram = trigram_model.generate(num_words=20, seed_text="Alice was")
    print(f"Seed: 'Alice was'")
    print(f"Output: {generated_trigram}\n")
    
    # 3. Quadgram Model (n=4)
    # The context length is n-1 = 3 words
    print("--- 4-Gram (Quadgram) Model ---")
    quadgram_model = NGramModel(n=4)
    quadgram_model.train(sample_corpus)
    quadgram_model.print_dictionary()
    generated_quadgram = quadgram_model.generate(num_words=20, seed_text="Alice was beginning")
    print(f"Seed: 'Alice was beginning'")
    print(f"Output: {generated_quadgram}\n")
```

## Evaluating Language Models: Perplexity

## What is Perplexity?

**Perplexity** is a mathematical measurement of how "confused" or "surprised" our Language Model is by a sequence of words. 
* A **lower** perplexity means the model finds the sentence highly predictable and natural (it expected it!).
* A **higher** perplexity means the model finds the sentence strange or improbable.

In the next cell, we are going to calculate the Perplexity of three different models (Bigram, Trigram, and Quadgram) against two different sentences:
1. A **"Seen"** sentence that actually exists in our training corpus.
2. An **"Unseen"** sentence that is completely new.

Let's see what happens to the perplexity as we increase *n*!


```python
# ==========================================
# CELL: Evaluating N-Gram Perplexity
# ==========================================

print("=== Evaluating Perplexity ===")
print("Perplexity measures how 'confused' the model is by the text.")
print("A LOWER score is better (the model predicted the sentence easily).\n")

# 1. We test on a completely "Seen" sentence (a sentence that was in the training corpus)
# This will show how larger N-grams are great at memorizing exact phrases.
test_sentence = "Alice was considering in her own mind whether the pleasure of making a daisy-chain would be worth the trouble."

# 2. We test on an "Unseen" sentence
# This will show how larger N-grams explode in perplexity when they encounter unfamiliar sequences.
unseen_sentence = "Alice was eating a pizza while driving a car."

print(f"Test Sentence (Seen): '{test_sentence}'")
print("-" * 50)
print(f"  Bigram Perplexity:   {bigram_model.calculate_perplexity(test_sentence):.2f}")
print(f"  Trigram Perplexity:  {trigram_model.calculate_perplexity(test_sentence):.2f}")
print(f"  Quadgram Perplexity: {quadgram_model.calculate_perplexity(test_sentence):.2f}")
print("  Insight: Notice how larger n-grams memorize the training text better, reaching lower perplexity!\n")

print(f"Test Sentence (Unseen): '{unseen_sentence}'")
print("-" * 50)
print(f"  Bigram Perplexity:   {bigram_model.calculate_perplexity(unseen_sentence):.2f}")
print(f"  Trigram Perplexity:  {trigram_model.calculate_perplexity(unseen_sentence):.2f}")
print(f"  Quadgram Perplexity: {quadgram_model.calculate_perplexity(unseen_sentence):.2f}")
print("  Insight: Notice how larger n-grams struggle (higher perplexity) with new text because they rely on long exact matches!")
```

## Analyzing the Results

### 1. The "Seen" Sentence
When evaluating the sentence from our training text, you should notice that **as `n` increases, the perplexity drops significantly.** 
* **Why?** A Quadgram looks at a 3-word history (context). In a small corpus, a specific 3-word context usually corresponds to *exactly one* possible next word. The model isn't "guessing" anymore; it has effectively **memorized** the text. It has absolute certainty, resulting in extremely low perplexity!

### 2. The "Unseen" Sentence
When evaluating a completely new sentence, the perplexity is uniformly much higher across the board. But critically, **larger n-grams struggle the most.**
* **Why?** Generalization. While a Bigram model might have seen transitions like "eating -> a" or "driving -> a" somewhere before, a Quadgram model demands an exact 3-word match. If it has never seen "eating a pizza" before, it assigns it a probability of near zero (thankfully saved by Laplace Smoothing). 

### The Core Trade-off of N-Grams
This perfectly illustrates the biggest challenge in traditional language modeling:
* **Large ** `n`: Great at generating realistic long sequences, but requires massive datasets to avoid memorizing (overfitting) and failing on new text.
* **Small ** `n`: Good at generalizing to new text, but generates grammatically unstructured "word salad" because it lacks long-term memory.
